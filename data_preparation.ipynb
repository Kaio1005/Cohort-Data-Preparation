```python
import pandas as pd
```

```python
ecommerce = pd.read_csv('data.csv', encoding='latin-1')
ecommerce.head()
```

```python
ecommerce.info()
```

```python
ecommerce.InvoiceDate = pd.to_datetime(ecommerce.InvoiceDate)
ecommerce.info()
```

```python
ecommerce.head()
```

```python
ecommerce['Cohort'] = ecommerce.InvoiceDate.dt.to_period('M').dt.to_timestamp('D')
ecommerce.info()
```

```python
ecommerce.head()
```

```python
ecommerce['LastOrder'] = ecommerce.groupby('CustomerID')['InvoiceDate'].transform('max')
ecommerce.head()
```

```python
ecommerce.info()
```

```python
ecommerce['LastCohort'] = ecommerce.Cohort.max()
ecommerce.head()
```

```python
ecommerce.to_csv('processed_data.csv', index = False)
```
